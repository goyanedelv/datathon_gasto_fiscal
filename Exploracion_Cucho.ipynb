```python
import pandas as pd
import matplotlib.pyplot as plt, mpld3
import numpy as np
import seaborn as sns

path_01="Core_Dataset_1_monthly_spend_2009-2017"
path_02="Core_Dataset_2_Anual_revenue_and_spend_2009-2017_by_Programs_(millions)"
path_03="Core_Dataset_3_Revenue_and_Spending_CG_Real_values"
path_04="Additional_Dataset_4_Population_of_Chile_1990-2017"
path_05="Additional_Dataset_5_GDP_Chilean_1990-2017_millions_chilean_pesos"
path_06="Additional_Dataset_6_Revenue_and_Spending_CG_Nominal values"

## Filenames
file_01="dataset_1_monthly_spend_2009-2017.csv"
file_02="dataset_2_Anual_revenue and spend_2009-2017 by Programs (millions).csv"
file_03="dataset_3_Revenue_and_ pending_(Central_Government)_1990-2017 (real).csv"
file_04="dataset_4_Population of Chile 1990 -2017.csv"
file_05="dataset_5_GDP Chilean 1990-2017_millions chilean pesos.csv"
file_06="dataset_6_Revenue_and_ pending_(Central_Government)_1990-2017 (nominal).csv"

df_cobre = pd.read_csv("../04_intermediate/ingresos_cobre.csv", names=['year','label','ing','cat','monto'], delimiter=";")
df_deficit = pd.read_csv("../04_intermediate/deficit_fiscal.csv")
df_deficit.rename(columns = {'TIME':'year', 'Value':'deficit'}, inplace = True)
df_deficit
```

```python
df_cobre['monto'].dtype.name
```

```python
df_cobre = df_cobre[df_cobre['cat']=='Cobre bruto']
if df_cobre['monto'].dtype.name == 'object':
    df_cobre['monto'] = df_cobre['monto'].str.replace(',','.').astype(float)
df_cobre['crecimiento'] = df_cobre['monto'].pct_change()
df_cobre
```

```python
year = 1990
presidential_years = [(2006,"Bachelet 1"), (2010, "PiÃ±era 1"), (2014, "Bachelet 2"), (2018, "")]

style_label = "seaborn-deep"
plt.rcParams["font.family"] = "Arial"

SMALL_SIZE = 10
MEDIUM_SIZE = 12
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('axes', titlesize=BIGGER_SIZE)  # fontsize of the figure title

with plt.style.context(style_label):
    fig, ax1 = plt.subplots(ncols=1, nrows=1, num=1,
                                 figsize=(10,5), squeeze=True, frameon=False)
    df_plot_cobre = df_cobre[df_cobre['year']>year]
    ax1.plot(df_plot_cobre['year'], df_plot_cobre['monto']/1000, marker='o', alpha=0.8);
    ax1.set_ylabel("Miles de millones de pesos")

    for i, (pres_year, pres_string) in enumerate(presidential_years):
        ax1.axvline(pres_year, color='gray', alpha=0.2)
        
        if pres_string:
            ax1.text((pres_year+presidential_years[i+1][0]-2.2)/2, 6000, color='grey', s=pres_string, alpha=1)
    
    sns.despine(right=True)
    ax1.set_title("Deficit fiscal e ingresos brutos del cobre")

    ax1.set_xlabel("AÃ±o")
    
    ax2 = ax1.twinx()
    anios_deficit = df_deficit[(df_deficit['LOCATION']=="CHL")]['year']
    monto_deficit = df_deficit[df_deficit['LOCATION']=="CHL"]['deficit']
    ax2.plot(anios_deficit,monto_deficit , 'r.')
    ax2.set_ylabel('Deficit/Superhabit fiscal', color='r', alpha=0.8)

#from matplotlib import savefig
plt.savefig('../05_output/cobre_y_deficit.png')

```
